# Transformada de Fourier con muestreo no uniforme

Cuando una señal no está uniformemente muestreada (los puntos no están espaciados regularmente en el tiempo), la transformada de Fourier tradicional (FFT) no se puede usar directamente.

### Opciones para el análisis de Fourier en este caso:

- **NDFT (Non-uniform Discrete Fourier Transform):** calcula la transformada considerando los tiempos reales de muestreo.
- **Interpolación:** transforma la señal a muestreo uniforme y luego aplica FFT.
- **NUFFT:** versión optimizada del NDFT.
- **Lomb-Scargle Periodogram:** útil cuando los datos son irregulares y se desea encontrar componentes periódicas.

A continuación, se muestra un ejemplo en Python usando NDFT.


```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una señal senoidal con muestreo no uniforme
np.random.seed(0)
t = np.sort(np.random.uniform(0, 1, 100))  # tiempos no uniformes
f0 = 5  # frecuencia en Hz
x = np.sin(2 * np.pi * f0 * t)

# Frecuencias para evaluar la NDFT
frequencies = np.linspace(0, 20, 1000)
X = np.array([np.sum(x * np.exp(-2j * np.pi * f * t)) for f in frequencies])

# Graficar la señal original
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(t, x, 'o-')
plt.title("Señal con muestreo no uniforme")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")

# Graficar el espectro
plt.subplot(1, 2, 2)
plt.plot(frequencies, np.abs(X))
plt.title("Espectro calculado con NDFT")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.tight_layout()
plt.show()
```

El análisis espectral de una señal con muestreo no uniforme se puede hacer utilizando la Transformada Discreta de Fourier No Uniforme (NDFT). A diferencia de la FFT, que requiere que las muestras estén uniformemente distribuidas, la NDFT toma en cuenta los tiempos exactos de muestreo. En el gráfico se observa una señal senoidal muestreada de manera irregular, y su espectro presenta un pico claro en 5 Hz, que corresponde a la frecuencia fundamental de la señal.
