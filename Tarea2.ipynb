```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import rfft, rfftfreq

# Parámetros de la señal
fs = 1000          # Frecuencia de muestreo [Hz]
T = 10             # Duración total [s]
t = np.linspace(0, T, fs*T, endpoint=False)  # Vector de tiempo

f_start = 1        # Frecuencia inicial [Hz]
f_end = 50         # Frecuencia final [Hz]

# Señal senoidal con frecuencia creciente (chirp lineal)
fase = 2 * np.pi * (f_start * t + ((f_end - f_start) / (2 * T)) * t**2)
s = np.sin(fase)

# Graficar la señal
plt.figure(figsize=(14, 4))
plt.plot(t, s)
plt.title("Señal senoidal con frecuencia creciente")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid()
plt.tight_layout()
plt.show()

# Transformada de Fourier
S = np.abs(rfft(s))
f = rfftfreq(len(t), 1/fs)

# Graficar la transformada
plt.figure(figsize=(10, 4))
plt.stem(f, S, basefmt=" ")
plt.title("Transformada de Fourier de la señal")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")
plt.xlim(0, 100)
plt.grid()
plt.tight_layout()
plt.show()

```
